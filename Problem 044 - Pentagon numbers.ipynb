Pentagon numbers

Problem 44

Pentagonal numbers are generated by the formula, Pn=n(3n1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 - 22 = 48, is not pentagonal.

<font color="red">Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk - Pj| is minimised; what is the value of D?</font>

```python
def gen_pentagon():
    current = 1
    while True:
        yield (current * ( (3*current) - 1 )) / 2.0
        current += 1
        
g = gen_pentagon()
for _ in range(10):
    print(g.next())
```

```python
class Pentagonal:
    "Class to check whether a number is pentagonal, with helper functions to generate pentagonal numbers as needed"
    def __init__(self):
        self.generator = gen_pentagon()
        self.numbers = set()
        self.numbers.add(self.generator.next())
        
    def is_pentagonal(self, n):
        if n > max(self.numbers):
            self.gen_until(n)
        return n in self.numbers
        
    def gen_until(self, n):
        while True:
            generated = self.generator.next()
            self.numbers.add(generated)
            if generated > n:
                break
        
    def __str__(self):
        nums = sorted(self.numbers)
        return "%s numbers between %s - %s" % (len(nums), nums[0], nums[-1])
    
p = Pentagonal()
print(p.is_pentagonal(48))
print(p)
```
