Highly divisible triangular number

Problem 12

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1

 3: 1,3

 6: 1,2,3,6

10: 1,2,5,10

15: 1,3,5,15

21: 1,3,7,21

28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

<font color="red">What is the value of the first triangle number to have over five hundred divisors?</font>

```python
def triangle_generator(n):
    return sum(xrange(n+1))

for i in range(1, 8):
    print(triangle_generator(i))
```

```python
def find_factors(n):
    import math
    results = set()
    results.add(1)
    results.add(n)
    for i in xrange(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            results.add(n / i)
            results.add(i)
    return results

for i in range(1, 8):
    tri = triangle_generator(i)
    print(i, tri, find_factors(tri))
```

```python
n = 1
while True:
    tri = triangle_generator(n)
    factors = find_factors(tri)
    if len(factors ) > 500:
        print(tri)
        break
    n += 1
        
```
