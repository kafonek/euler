Cyclical figurate numbers

Problem 61

Triangle, square, pentagonal, hexagonal, heptagonal, and octagonal numbers are all figurate (polygonal) numbers and are generated by the following formulae:

Triangle	 	P3,n=n*(n+1)/2	 	1, 3, 6, 10, 15, ...

Square	 	P4,n=n^2	 	1, 4, 9, 16, 25, ...

Pentagonal	 	P5,n=n*(3n-1)/2	 	1, 5, 12, 22, 35, ...

Hexagonal	 	P6,n=n*(2n-1)	 	1, 6, 15, 28, 45, ...

Heptagonal	 	P7,n=n*(5n-3)/2	 	1, 7, 18, 34, 55, ...

Octagonal	 	P8,n=n*(3n-2)	 	1, 8, 21, 40, 65, ...

The ordered set of three 4-digit numbers: 8128, 2882, 8281, has three interesting properties.

1. The set is cyclic, in that the last two digits of each number is the first two digits of the next number (including the last number with the first).

2. Each polygonal type: triangle (P3,127=8128), square (P4,91=8281), and pentagonal (P5,44=2882), is represented by a different number in the set.

3. This is the only set of 4-digit numbers with this property.

<font color="red">Find the sum of the only ordered set of six cyclic 4-digit numbers for which each polygonal type: triangle, square, pentagonal, hexagonal, heptagonal, and octagonal, is represented by a different number in the set.</font>

```python
import euler
from itertools import dropwhile, takewhile
### Using the generators written in euler.py along with itertools dropwhile and takewhile, 
### it becomes possible to write "return all numbers in a generator between 1,000-10,000"
### or similar conditions.
```

```python
nums = []
for i, generator in enumerate([euler.triangle(), euler.square(), euler.pentagonal(), euler.hexagonal(), euler.heptagonal(), euler.octagonal()]):
    nums.extend([(i+3, str(t)) for t in dropwhile(lambda x: x < 1000, takewhile(lambda y: y < 10000, generator))])
len(nums)
    
```

```python
def expand(chain, checked=[]):
    'Recursive approach to creating possible chains and eliminating them'
    checked.append(chain)
    if not chain:
        return False
    if len(chain) == 6:
        #check if the first number in the chain is cyclical with the last number in the chain
        if chain[0][1][:2] == chain[-1][1][2:]:
            return chain
        else:
            checked.append(chain)
            return expand(chain[:-1], checked)
    types = [tup[0] for tup in chain]
    prefix = chain[-1][1][2:] #last two numbers in the string of the last number in the chain
    for num in nums:
        if num[1][:2] == prefix:
            if num[0] not in types:
                if chain + [num] not in checked:
                    chain.append(num)
                    return expand(chain, checked)
    #no candidate numbers, drop the last number in the chain and recurse
    chain = chain[:-1]
    return expand(chain, checked)
```

```python
for num in nums:
    start_chain = [num]
    answer = expand(start_chain)
    if answer:
        print(answer)
        print(sum(int(x[1]) for x in answer))
        break
```
