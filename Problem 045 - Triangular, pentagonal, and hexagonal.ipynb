Triangular, pentagonal, and hexagonal

Problem 45

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...

Pentagonal	 	Pn=n(3n-1)/2	 	1, 5, 12, 22, 35, ...

Hexagonal	 	Hn=n(2n-1)	 	1, 6, 15, 28, 45, ...

It can be verified that T285 = P165 = H143 = 40755.

<font color="red">Find the next triangle number that is also pentagonal and hexagonal.</font>

```python
def gen_numbers():
    current = 1
    while True:
        triangle = (current * (current + 1)) / 2
        pentagonal = (current * ( (3*current) - 1) ) / 2
        hexagonal = current * ( (2*current) - 1) 
        yield triangle, pentagonal, hexagonal
        current += 1
        
g = gen_numbers()
for _ in range(5):
    print(g.next())
```

```python
pents = set()
hexes = set()
g = gen_numbers()
for _ in range(300):
    tri, pent, hex = g.next()
    pents.add(pent)
    hexes.add(hex)
    if tri in pents and tri in hexes:
        print(tri)
```
